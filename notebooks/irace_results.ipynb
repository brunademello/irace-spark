```python
# Dados de exemplo
config = ['1', '2', '3', '4']
mean = [176.5962022, 176.00921, 176.1499423, 176.537801]

config_spark = ['1', '2', '3', '4']
mean_spark = [198.5100105, 198.3582154, 198.323243, 199.0098858]

config_cassandra = ['1', '2', '3', '4']
mean_cassandra = [492.2333869, 494.9429938, 498.9875978, 496.5151548]

# Definindo a largura das barras
bar_width = 0.2

# Definindo as posi√ß√µes das barras manualmente para evitar sobreposi√ß√£o
position_config = [i - bar_width for i in range(1, 5)]
position_config_spark = [i for i in range(1, 5)]
position_config_cassandra = [i + bar_width for i in range(1, 5)]

# Criando o gr√°fico de barras com as barras lado a lado
plt.bar(position_config_cassandra, mean_cassandra, color='0.7', width=bar_width, label='Cassandra')
plt.bar(position_config_spark, mean_spark, color='0.4', width=bar_width, label='Spark')
plt.bar(position_config, mean, color='0.1', width=bar_width, label='Outros')

plt.xlabel('Configura√ß√£o')
plt.ylabel('M√©dia (s)')

# Adicionando legenda
plt.legend()

plt.show()
```
